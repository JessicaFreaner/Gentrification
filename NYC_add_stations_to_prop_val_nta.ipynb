# LOAD - NTA w/ PROPERTY VALUES

```python
manh_nta_property_values_df = pickle.load( open( "manh_nta_property_values.pkl", "rb" ) )
bk_nta_property_values_df = pickle.load( open( "bk_nta_property_values.pkl", "rb" ) )
qu_nta_property_values_df = pickle.load( open( "qu_nta_property_values.pkl", "rb" ) )
bx_nta_property_values_df = pickle.load( open( "bx_nta_property_values.pkl", "rb" ) )
st_nta_property_values_df = pickle.load( open( "st_nta_property_values.pkl", "rb" ) )
```

```python
boro_df_list = [manh_nta_property_values_df, 
                bk_nta_property_values_df, 
                qu_nta_property_values_df, 
                bx_nta_property_values_df,
                st_nta_property_values_df]
```

```python
nyc_nta_property_values_df = pd.concat(boro_df_list, axis=0)
```

# ADD Prop Values to Station ID 

```python
geocoded_station_nta_df.tail()
```

```python
nyc_nta_property_values_df.head()
```

```python
nyc_nta_property_values_df.rename(columns={'name': 'NTAName'}, inplace=True)
```

```python
nyc_nta_property_values_df.head()
```

```python
grouped_median_nyc_nta_df = nyc_nta_property_values_df.groupby('NTAName').median() 
```

```python
grouped_median_nyc_nta_df
```

```python
geocoded_nyc_prop_values_df = geocoded_station_nta_df.join(grouped_median_nyc_nta_df, on='NTAName')
```

```python
geocoded_nyc_prop_values_df
```

```python

with open("geocoded_station_and_nta_prop_values.pkl", "w") as picklefile:
        pickle.dump( geocoded_nyc_prop_values_df, picklefile )
```
