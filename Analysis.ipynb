```python
def plot_yr_prop_values_and_prev_yr(df, xyear, yyear, mode='default'):
    X = df[[xyear]]
    X = sm.add_constant(X)
    if mode == 'default':
        Y = df[yyear] 
    elif mode == 'diff':
        Y = df[yyear] - df[xyear]
    elif mode == 'pct_diff':
        Y = ( df[yyear] - df[xyear] ) / df[xyear]
#     elif mode == 'log':
#         X =  df[[xyear]].apply(np.log)
#         X = sm.add_constant(X)
#         Y = df[yyear].apply(np.log)
 
    linmodel = sm.OLS(Y, X).fit()

    plt.scatter(df[xyear], Y)
    plt.xlabel('%s Median Prices' % xyear)
    plt.ylabel('%s Median Prices' % xyear)
    plt.title("Prices Based on Last Year's Prices")
    if mode == 'default':
        plt.plot(df[xyear], linmodel.predict(), 'r')
    elif mode == 'diff':
        plt.plot(df[xyear], linmodel.predict(), 'c')
    elif mode == 'pct_diff':
        plt.plot(df[xyear], linmodel.predict(), 'm')

#     elif mode == 'log':
#         plt.plot(df[xyear], linmodel.predict(), 'g')
```

## Convert to DATETIME Objects

```python
def convert_datetime(dt_str):
    return datetime.strptime(dt_str, '%Y-%m-%d')

def convert_df_col_to_date(df, col='audit_date'):
    df[col] = df.apply(lambda x: convert_datetime(x[col]),
                      axis=1)
```
