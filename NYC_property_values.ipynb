### ## group by name (NTAName)

```python
def calc_nta_avg_prop_val(df, nta_name='name'):
    return df.groupby(nta_name).mean()
```

```python
grouped_manh_nta = calc_nta_avg_prop_val(manh_nta_property_values_df)
grouped_bk_nta = calc_nta_avg_prop_val(bk_nta_property_values_df)
grouped_qu_nta = calc_nta_avg_prop_val(qu_nta_property_values_df)
grouped_bx_nta = calc_nta_avg_prop_val(bx_nta_property_values_df)
grouped_st_nta = calc_nta_avg_prop_val(st_nta_property_values_df)
```

## Create Dict for Property Values

Match on data[ 'features' ][ 'properties' ][ 'NTAName' ] = df[ 'name' ]

```python
def make_property_value_dict(df):
    pv_dict = {}
    for i, index in enumerate(df.index):
        key = index
        values = {}
        for col in df.columns[:-1]:
            values[col] = df.iloc[i][col]
        pv_dict[index] = values
#         print pv_dict
#         break
    return pv_dict
```

```python
manh_val_dict = make_property_value_dict(grouped_manh_nta )
bk_val_dict = make_property_value_dict(grouped_bk_nta )
qu_val_dict = make_property_value_dict(grouped_qu_nta )
bx_val_dict = make_property_value_dict(grouped_bx_nta )
st_val_dict = make_property_value_dict(grouped_st_nta )
```
