```python
# determine if a point is inside a given polygon or not
# Polygon is a list of (x,y) pairs.

def point_inside_polygon(x,y,poly):

    n = len(poly)
    inside =False

    p1x,p1y = poly[0]
    for i in range(n+1):
        p2x,p2y = poly[i % n]
        if y > min(p1y,p2y):
            if y <= max(p1y,p2y):
                if x <= max(p1x,p2x):
                    if p1y != p2y:
                        xinters = (y-p1y)*(p2x-p1x)/(p2y-p1y)+p1x
                    if p1x == p2x or x <= xinters:
                        inside = not inside
        p1x,p1y = p2x,p2y

    return inside
```

```python
def extract_nta_info(geojson_dict):
    nta_dict = {}
    for feature in geojson_dict['features']:
        nta_code = feature['properties']['NTACode']
        nta_name = feature['properties']['NTAName']
        if feature['geometry']['type'] == 'Polygon':
            polygon_coor = feature['geometry']['coordinates'][0]
            nta_dict[nta_code] = {'NTAName': nta_name, 'NTACode': nta_code, 'polygon': polygon_coor}
        else:
            polygon_coor = feature['geometry']['coordinates']
            nta_dict[nta_code] = {'NTAName': nta_name, 'NTACode': nta_code, 'polygon': polygon_coor}
#         pprint(nta_dict)
#         break
    return nta_dict
```

```python
def get_nta_code(station_lng, station_lat, poly_dict):
    for k,v in poly_dict.items():

        try: # logic for Polygon
            if point_inside_polygon(station_lng, station_lat, poly_dict[k]['polygon']):
#                 print '\n FOUND!!!! NTA ', k
                return k
            else:
                continue
        except: # logic for MultiPolygon
#             print "FAILED to check", k
            for i, poly in enumerate(poly_dict[k]['polygon']):
                if point_inside_polygon(station_lng, station_lat, poly_dict[k]['polygon'][i][0]):
#                     print '\n FOUND!!!! NTA ', k
                    return k
                else:
                    continue
```

```python
def get_nta_name(station_lng, station_lat, poly_dict):
    for k,v in poly_dict.items():

        try: # logic for Polygon
            if point_inside_polygon(station_lng, station_lat, poly_dict[k]['polygon']):
                return poly_dict[k]['NTAName']
            else:
                continue
        except: # logic for MultiPolygon
            for i, poly in enumerate(poly_dict[k]['polygon']):
                if point_inside_polygon(station_lng, station_lat, poly_dict[k]['polygon'][i][0]):
                    return poly_dict[k]['NTAName']
                else:
                    continue
```
