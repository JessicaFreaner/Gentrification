```python
from pprint import pprint
import urllib2
from bs4 import BeautifulSoup
import dateutil.parser
import re
import pandas as pd
import numpy as np
import csv
import pickle

import os
```

```python
# extract soup for urlname
def url_to_soup(urlname):
    try:
        page = urllib2.urlopen(urlname)
        soup = BeautifulSoup(page)
    except:
        print urlname
        soup = 'N/A'
        print 'soup', soup
    return soup
```

```python
def find_urls(soup, identifier):
    try:
        return soup.find_all('a', href=re.compile(identifier))
    except:
        print 'failed soup', soup
```

```python
def extract_data_urls(urlname):
    soup = url_to_soup(urlname)
    data_urls = []
    
    # identify url for weekly data
    link_list = find_urls(soup, 'data/nyct/turnstile/')
    for link in link_list:
        one_link = "http://web.mta.info/developers/" + link.get('href')
        data_urls.append(str(one_link))
#     pp.pprint(data_urls)
    return data_urls
```

```python
def wget_urls(url_list):
    for url in url_list:
        os.system('wget %s' %url)
```

```python
def mta_turnstile_data_extraction():
    start_url = "http://web.mta.info/developers/turnstile.html"
    
    # main page urls
    weekly_data_urls = extract_data_urls(start_url)
    print '%i URLS extracted' % len(weekly_data_urls)
    
    # wget to retrieve data from url to local file
#     wget_urls(weekly_data_urls)
    return weekly_data_urls
```

```python
def ts_function(mta_dict, turnstile_file, end_of_key_index=5):
    # Open up a file
#     with open('turnstile_150110.txt', 'r') as turnstile_file:
    with open(turnstile_file, 'r') as turnstile_file:

        mta_values = []
    # use csv reader to read it
        reader = csv.reader(turnstile_file)
        for count, row in enumerate(reader):
        # make dict with key for each (C/A, UNIT, SCP, STATION, LINENAME)
        # do NOT ditinguish between turnstiles
            if count == 0:
                continue # skip 1st row of headers
            
            mta_key = tuple(row[:end_of_key_index])
            mta_values = row[end_of_key_index:]
        
            # check if key exists in dict 
            if mta_key not in mta_dict:
                mta_dict[mta_key] = []
                
            mta_dict[mta_key].append(mta_values)
            
    turnstile_file.close()
        
    return mta_dict
```

```python
def from_url_to_file(urls, name_length=20, key_length=5, newer=True):
    ts_dict = {}
    for url in urls:
        fname = url[-name_length:] # get name from url ending
        if newer:
            ts_dict = ts_function(ts_dict, fname, key_length)
        else:
            ts_dict = ts_function2(ts_dict, fname)
    return ts_dict
```
